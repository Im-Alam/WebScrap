```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
home_url = 'https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget'
url_p2 = 'https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page=2'
header = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/112.0'}

flag = True
count = 1

name = []
rating = []
high_rating_values = []
bad_rating_factor = []
Review = []
Salary_comments = []
Interview_Q = []
Jobs = []
Job_link = []
Photos = []
Sector = []
Employee_count = []
Ownership = []
Age = []

while flag:
    try:
        if count == 1:
            url = home_url
            count += 1
        elif count == 2:
            url = url_p2
            count += 1
        else:
            url = url_p2[0:-1] + f'{count}'
            count+=1
        print(count)
        webpage = requests.get(url, headers=header).text
        soup = BeautifulSoup(webpage, 'html5')
        companies = soup.find_all(class_ = 'companyCardWrapper')
        
        for company in companies:
            try:
                name.append(company.find(class_='companyCardWrapper__companyName').text.strip())
            except AttributeError:
                name.append('NA')

            # Rating
            try:
                rating.append(company.find(class_='companyCardWrapper__companyRatingValue').text.strip())
            except AttributeError:
                rating.append('NA')

            # High rating values
            try:
                high_rating_values.append(company.find_all(class_='companyCardWrapper__ratingValues')[0].text)
            except (AttributeError, IndexError):
                high_rating_values.append('NA')

            # Bad rating factor
            try:
                bad_rating_factor.append(company.find_all(class_='companyCardWrapper__ratingValues')[1].text)
            except (AttributeError, IndexError):
                bad_rating_factor.append('NA')

            # ActionCount
            ActionCount = [count.text if count else 'NA' for count in company.find_all(class_="companyCardWrapper__ActionCount")]

            # Review
            try:
                Review.append(ActionCount[0])
            except IndexError:
                Review.append('NA')

            # Salary comments
            try:
                Salary_comments.append(ActionCount[1])
            except IndexError:
                Salary_comments.append('NA')

            # Interview Questions
            try:
                Interview_Q.append(company.find_all('a', class_="companyCardWrapper__ActionWrapper")[2]['href'])
            except (AttributeError, IndexError):
                Interview_Q.append('NA')

            # Jobs
            try:
                Jobs.append(ActionCount[3])
            except IndexError:
                Jobs.append('NA')

            # Job link
            try:
                Job_link.append(company.find_all('a', class_="companyCardWrapper__ActionWrapper")[3]['href'])
            except (AttributeError, IndexError):
                Job_link.append('NA')

            # Photos
            try:
                Photos.append(company.find_all('a', class_="companyCardWrapper__ActionWrapper")[5]['href'])
            except (AttributeError, IndexError):
                Photos.append('NA')

            # Sector
            try:
                Sector.append(company.find(class_='companyCardWrapper__interLinking').text.strip().split('|')[0])
            except (AttributeError, IndexError):
                Sector.append('NA')

            # Employee count
            try:
                Employee_count.append(company.find(class_='companyCardWrapper__interLinking').text.strip().split('|')[1])
            except (AttributeError, IndexError):
                Employee_count.append('NA')

            # Ownership
            try:
                Ownership.append(company.find(class_='companyCardWrapper__interLinking').text.strip().split('|')[2])
            except (AttributeError, IndexError):
                Ownership.append('NA')

            # Age
            try:
                Age.append(company.find(class_='companyCardWrapper__interLinking').text.strip().split('|')[-2])
            except (AttributeError, IndexError):
                Age.append('NA')
            
            if count == 50
                flag = False

    except:
        flag = False
        
```

```python
Columns = ["name", "rating", "high_rating_values", "bad_rating_factor", "more", "Review", "Salary_comments", "Interview_Q", "Jobs", "Job_link", "Photos", "Sector", "Employee_count", "Ownership", "Age"]
data = {
    "name": name,
    "rating": rating,
    "Sector": Sector,
    "Ownership": Ownership,
    "high_rating_values": high_rating_values,
    "bad_rating_factor": bad_rating_factor,
    "Review": Review,
    "Salary_comments": Salary_comments,
    "Interview_Q": Interview_Q,
    "Jobs": Jobs,
    "Job_link": Job_link,
    "Photos": Photos,
    "Employee_count": Employee_count,
    "Age":Age
}
Data = pd.DataFrame(data)
Data.to_csv('ambitioBox2.csv')
```
